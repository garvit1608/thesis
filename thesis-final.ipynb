```python
#Gender transition matrix

#Male

p_transition_male = np.array(
    [[0.6779, 0.1370, 0.0867, 0.0181, 0.0800],   #Working
     [0, 1, 0, 0, 0],#Resign
     [0, 0, 1, 0, 0],#Layoff
     [0, 0, 0, 1, 0],#Retired
     [0, 0, 0, 0, 1],#Promotion
    ]
)

#Female

p_transition_female = np.array(
    [[0.4746, 0.2068, 0.1385, 0.0248, 0.1551],   #Working
     [0, 1, 0, 0, 0],#Resign
     [0, 0, 1, 0, 0],#Layoff
     [0, 0, 0, 1, 0],#Retired
     [0, 0, 0, 0, 1],#Promotion
    ]
)
```

```python
def get_age_transition_matrix(age):
    if age >= 26 and age <= 40:
        return p_transition_agegp1
    elif age >= 41 and age <= 60:
         return p_transition_agegp2
    else:
         return p_transition_agegp3
    
def get_dept_transition_matrix(dept_id):
    if dept_id == 1:
          return p_transition_admin
    elif dept_id == 2:
          return p_transition_exec
    elif dept_id == 3:
          return p_transition_its
    elif dept_id == 4:
          return p_transition_sw
    elif dept_id == 5:
          return p_transition_prod
    else:
        return p_transition_sales
```

```python
p_init = np.array([1, 0, 0, 0, 0])
states = ['Working', 'Resign', 'Layoff', 'Retire', 'Promotion']
p_state_t = [p_init]


# p_transition = get_age_transition_matrix(row['Age']) @ get_dept_transition_matrix(row['DeptID'])
p_transition = get_dept_transition_matrix(4)
# p_transition = get_age_transition_matrix(row['Age']) 
# p_transition = p_transition_agegp3


for i in range(10):  # 200 time steps sorta, kinda, approximates infinite time :)
    p_state_t.append(p_state_t[-1] @ p_transition)

for state in states:
    print(state, end=' ')
print('')
for i in range(11):
    for j in range(5):
        print(float(f'{p_state_t[i][j]:.3f}'), end='\t')
    print('')
# p_state_t
state_distributions = pd.DataFrame(p_state_t)
# state_distributions.plot()
plt.plot(state_distributions)
plt.ylabel('Probability')
plt.xlabel('Year')
plt.title('Software Engineering')
plt.legend(['Working', 'Resign', 'Layoff', 'Retire', 'Promotion'])
plt.show()
```
